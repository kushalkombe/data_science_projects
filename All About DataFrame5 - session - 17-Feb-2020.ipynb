```python
# Sort on multiple columns
df.sort_values(by='col6', na_position='first') # bydeafult all NAns are placed at last 
```

```python
df.sort_values(by='col2', kind='mergesort')  #default sort algorithm
```

```python
# This is depricated method
# by default it picked up col1 as an index and sorted, here we are sorting in descending order
df1.sort_index(axis=0,
    level=None,
    ascending=True,
    inplace=False,
    kind='quicksort',
    na_position='last',
    sort_remaining=True,
    by=None)
```

```python
# This is depricated method
# by default it picked up col1 as an index and sorted, here we are sorting in descending order
df1.sort_index(axis=0,
    level=None,
    ascending=[True,False],
    inplace=False,
    kind='quicksort',
    na_position='last',
    sort_remaining=True,
    by=['col2','col1'])
```
